```
import time
from random import randint
from random import uniform
import copy
import math
import numpy as np
```

```
def RandomIniSolutionGenerate(ebees):
    neighbours=[]
    a=0;b=2;c=0;d=3
    for e in range(ebees):
        sudoku=[[3,5,0,0,4,0],[0,0,1,0,0,0],[6,0,0,0,1,0],[0,0,5,3,0,0],[0,0,3,4,0,2],[0,0,0,0,3,0]] #hard
        #sudoku=[[0,4,0,2,0,0],[2,0,5,0,0,0],[0,6,1,0,0,0],[0,0,0,4,6,0],[0,0,0,5,0,3],[0,0,3,0,1,0]] #easy
        #sudoku=[[0,0,3,6,0,0],[0,2,0,0,0,4],[5,0,0,0,6,0],[0,3,0,0,0,5],[3,0,0,0,1,0],[0,0,1,4,0,0]]  #easy
        #sudoku=[[1,6,0,0,0,5],[0,0,5,2,0,0],[5,0,0,0,3,0],[0,4,0,0,0,1],[0,0,4,1,0,0],[3,0,0,0,5,4]]  #easy
        a=0;b=2;c=0;d=3
        for count in range(len(sudoku)):
            #print("ITERATION",count)
            value=[]
            for i in range(a,b):
                for j in range(c,d):
                 #  print("a,b,c,d",a,b,c,d)
                    value.append(sudoku[i][j])
            for i in range(a,b):
                for j in range(c,d):
                    if sudoku[i][j]==0:
                        rno=randint(1,6)
                        if(rno in value):
                            while(rno in value):
                                rno=randint(1,6)
                            sudoku[i][j]=rno
                            value.append(rno)
                        else:
                            sudoku[i][j]=rno
                            value.append(rno)
            if(count==0):
                c+=3;d+=3
            if(count==1):
                a+=2;b+=2
            if(count==2):
                c-=3;d-=3
            if(count==3):
                a=4;b=6
            if(count==4):
                c+=3; d+=3
        neighbours.append(sudoku)
    return neighbours
```

```
def FitnessCalculation(beeno,neighbours):
    '''counting missing values in row for penalty value calculation'''
    #print("NEIGHBOURS",neighbours)
    unique=[1,2,3,4,5,6]
    row_count=[]

    for k in range(len(neighbours)):
        ct=0
        for i in range(len(neighbours[0])):
            for j in range(len(unique)):
                #print("final_row[k][i]",neighbours[k][i])
                if unique[j] not in neighbours[k][i]:
                    ct+=1
        #print("COUNT",k,ct)
        row_count.append(ct)
    #print("ROW COUNT OF ", beeno,"is",row_count[beeno])


    '''row to column conversion'''
    final_col=[]
    for k in range(len(neighbours)):
        col=[]
        for i in range(len(neighbours[0])):
            #print("neighbour[i]",neighbours[i])
            col_list=[]
            for m in range(len(neighbours[0])):
                col_list.append(neighbours[k][m][i])
            #print("COL_LIST",col_list)
            col.append(col_list)
        final_col.append(col)

    col_count=[]
    '''calculating missing values in column for penalty value calculation'''
    for k in range(len(neighbours)):
        col_ct=0
        for i in range(len(final_col[0])):
            for j in range(len(unique)):
                final_col[k][i]=list(final_col[k][i])
                if unique[j] not in final_col[k][i]:
                    col_ct+=1
        col_count.append(col_ct)
    #print("COL COUNT OF ", beeno,"is",col_count[beeno])

    '''calculating fitness,penalty,probability'''
    fitness=[]
    for i in range(len(row_count)):
        penalty=row_count[i]+col_count[i]
        fit=1/(1+penalty)
        fitness.append(fit)
    #print("FITNESS",fitness)
    prob=[]
    for i in range(len(fitness)):
        p=fitness[i]/sum(fitness)
        #print("p,fit[i]sumof_f",p,fitness[i],sum(fitness))
        prob.append(p)
    #print("PROBABILITY OF",beeno,"is",prob[beeno])
    return prob[beeno],fitness[beeno]
```

```
def SubgridConstrainCheck(array,valueofj,j):
    a=0;b=2;c=0;d=3
    rows=[1,2,3,4,5,6]
    true=[]
    for count in range(6):
        actual_row=[]
        for i in range(a,b):
            for k in range(c,d):
                actual_row.append(array[i][k])
        #print(actual_row)
        actual_row.sort()
        if(actual_row==rows):
            true.append(1)
        else:
            missing_val=set(rows).difference(actual_row)
            missing_val=list(missing_val)
            index =j
            numberOfColumns = len(sudoku)
            column = index%numberOfColumns
            row    = math.floor(index / numberOfColumns)
            for i in range(a,b):
                for k in range(c,d):
                    if (array[i][k] == valueofj and (i,k)!=(row,column)):
                      array[i][k]=missing_val[0]
        if(count==0):
          c+=3;d+=3
        if(count==1):
          a+=2;b+=2
        if(count==2):
          c-=3;d-=3
        if(count==3):
          a=4;b=6
        if(count==4):
          c+=3; d+=3
    return array
```

```
'''parameters initialisation'''
ebees=20
obees=40

'''getting the start square index values'''
#sudoku=[[1,6,0,0,0,5],[0,0,5,2,0,0],[5,0,0,0,3,0],[0,4,0,0,0,1],[0,0,4,1,0,0],[3,0,0,0,5,4]]
#sudoku=[[0,4,0,2,0,0],[2,0,5,0,0,0],[0,6,1,0,0,0],[0,0,0,4,6,0],[0,0,0,5,0,3],[0,0,3,0,1,0]]
#sudoku=[[0,0,3,6,0,0],[0,2,0,0,0,4],[5,0,0,0,6,0],[0,3,0,0,0,5],[3,0,0,0,1,0],[0,0,1,4,0,0]]
sudoku=[[3,5,0,0,4,0],[0,0,1,0,0,0],[6,0,0,0,1,0],[0,0,5,3,0,0],[0,0,3,4,0,2],[0,0,0,0,3,0]] #hard
sudoku=np.array(sudoku)
sudoku=sudoku.flatten()
start_square=[]
for i in range(len(sudoku)):
    if(sudoku[i]>0):
        start_square.append(i)
sudoku=sudoku.reshape(-1,6)

'''neighbours initialisation'''


neighbours= RandomIniSolutionGenerate(ebees)

'''neighbourhood search'''
'''solution search cycle'''
start=time.time()
cycle=0
max_cycle=100 #max_no_cycles
max_ff=0
while (cycle!=max_cycle and max_ff!=1):
    print("CYCLE",cycle)
    for bee in range(ebees):
        '''initial fitness calculation'''
        #print("neighbour[bee] before 1st fit function",neighbours[bee])
        prob,fit=FitnessCalculation(bee,neighbours)
        #print("neighbour[bee] after 1st fit function",neighbours[bee])
        '''calculation of number of onlooker bees for the food source'''
        onlooker=round(prob*obees)
        for b in range(onlooker+1):
            '''k value'''
            k1=randint(0,5)
            while(k1==bee):
               k1=randint(0,5)
            '''j value'''
            j=randint(0,35)
            while(j in start_square):
               j=randint(0,35)
            neighbours[bee]=np.array(neighbours[bee])
            neighbours[k1]=np.array(neighbours[k1])
            x=neighbours[bee].flatten()
            k=neighbours[k1].flatten()
            #print("x,k",x.reshape(-1,6),k.reshape(-1,6))
            v=copy.deepcopy(x)
            '''neighbourhood calculation formula'''
            v[j]=math.ceil(x[j]+uniform(0,1)*abs(x[j]-k[j]))
            '''checking v[j] is greater than n ie.6'''
            if(v[j] > len(sudoku)):
                v[j]=np.mod(v[j],len(sudoku))
            valueofj=v[j]
            v=v.reshape(-1,6)
            x=x.reshape(-1,6)
            #print("BEFORE V",v)
            '''checking for the subgrid constrain'''
            v=SubgridConstrainCheck(v,valueofj,j)
            #print("AFTER V",v)
            x=x.reshape(-1,6)
            neighbours.append(x)
            #print("neighbours",neighbours)
            '''fitness calculation of x'''
            probx,fitx=FitnessCalculation(len(neighbours)-1,neighbours)
            neighbours.pop(len(neighbours)-1)
            #print("FITNESS OF X",fitx)
            '''fitness calculation of v'''
            neighbours.append(v)
            probv,fitv=FitnessCalculation(len(neighbours)-1,neighbours)
            #print("FITNESS OF V",fitv)
            neighbours.pop(len(neighbours)-1)
            '''comparing the fitness and storing the best solution'''
            if fitv > fitx:
              neighbours[bee]=v


    fp=[];ff=[]
    '''finding best among the best solutions in employed bee and onlooker bee phase'''
    for i in range(len(neighbours)):
        finalprob,finalfit=FitnessCalculation(i,neighbours)
        if finalfit==1:
          print("Solution")
          print(neighbours[i])
        fp.append(finalprob)
        ff.append(finalfit)
    max_fp=max(fp)
    max_ff=max(ff)
    #print("FITNESS",ff)
    #print("PROB",fp)
    print("MAX PROB:",max_fp,"MAX FIT:",max_ff)


    '''Scout bee phase ->randomly finding solution through exploration'''
    ffp=[];fff=[]
    sbees=int(0.1*ebees)
    sorted_ff=sorted(ff)
    abondaned_fit=[]
    abondaned_fit_index=[]
    #print(sbees)
    for i in range(sbees):
      abondaned_fit.append(sorted_ff[i])
      countt=0
      for h in range(len(ff)):
        if (ff[h]== sorted_ff[i]):
          if(countt==0):
            abondaned_fit_index.append(h)
            #print(abondaned_fit_index)
            countt+=1
    abondaned_solution=[]
    for i in abondaned_fit_index:
      abondaned_solution.append(neighbours[i])
    #print(abondaned_solution)

    scout_neighbours= RandomIniSolutionGenerate(sbees)
    #print("SCOUT_NIEGHBOURS",scout_neighbours)
    for i in range(len(scout_neighbours)):
        finalprob_scout,finalfit_scout=FitnessCalculation(i,scout_neighbours)
        if finalfit_scout > abondaned_fit[i]:
          neighbours[i]=scout_neighbours[i]
        else:
          neighbours[i]=abondaned_solution[i]
    cycle+=1
end=time.time()
print(f"Runtime of the program is {end-start} seconds")

```
